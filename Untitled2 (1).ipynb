```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Instalar dependencias si es necesario
try:
    import lxml
except ImportError:
    !pip install lxml

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Método con lxml
try:
    tesla_revenue = pd.read_html(url, flavor='lxml')[0]
    tesla_revenue.columns = ['Date', 'Revenue']
    tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"", regex=True)
    tesla_revenue.dropna(inplace=True)
    tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
    print("Con lxml:")
    tesla_revenue.tail()
    
except Exception as e:
    print(f"Error con lxml: {e}")
    
    # Método alternativo con html5lib
    try:
        tesla_revenue = pd.read_html(url, flavor='html5lib')[0]
        tesla_revenue.columns = ['Date', 'Revenue']  
        tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"", regex=True)
        tesla_revenue.dropna(inplace=True)
        tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
        print("Con html5lib:")
        tesla_revenue.tail()
        
    except Exception as e2:
        print(f"Error con html5lib: {e2}")
        print("Usando método BeautifulSoup manual...")
```
